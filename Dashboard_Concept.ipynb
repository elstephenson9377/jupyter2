# Visualization Examples using pandas, matplotlib, numpy, seaborn

# Example 1 event type bar graph and pie chart combo

```python
# an example code of how to read/view data into notebooks

import pandas as pd
import matplotlib.pyplot as plt

# Load data from Excel file
df = pd.read_excel('example computer log file.xlsx')

# Create a bar chart of the count of events by type
plt.figure(figsize=(8,6))
df['event_type'].value_counts().plot(kind='bar', color='red')
plt.title('event_type Counts')
plt.xlabel('Event Type')
plt.ylabel('Count')
plt.show()

# Create a pie chart of the count of events by severity level
plt.figure(figsize=(8,6))
df['username'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90, counterclock=False)
plt.title('Event Counts')
plt.legend(loc='upper left', bbox_to_anchor=(1.0, 1.0))
plt.show()
```

# Example 2 Bar chart and Line Graph Combo

```python
#@title Default title text
# this is an example of # of events by user and hourly event counts

import pandas as pd
import matplotlib.pyplot as plt

# Load data from Excel file into pandas dataframe
data = pd.read_excel('example computer log file.xlsx')

# Count number of events by username
event_counts = data.groupby('username').count()['event_type']

# Create a bar chart of event counts by username
event_counts.plot(kind='bar')
plt.title('Number of Events by User')
plt.xlabel('Username')
plt.ylabel('Number of Events')
plt.show()

# Count number of events by hour of day
data['hour'] = data['timestamp'].dt.hour
hourly_counts = data.groupby('hour').count()['event_type']

# Create a line chart of hourly event counts
hourly_counts.plot()
plt.title('Hourly Event Counts')
plt.xlabel('Hour of Day')
plt.ylabel('Number of Events')
plt.show()
```

# Example 3 Heat Map Visualization 

```python
# this is a heatmap example using the same data

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Load the log data from an xlsx file
log_data = pd.read_excel('example computer log file.xlsx')

# Extract the relevant columns (timestamp, username, and event_type)
log_data = log_data[['timestamp', 'username', 'event_type']]

# Pivot the data to create a matrix with timestamps as rows, usernames as columns, and event_types as values
log_matrix = log_data.pivot_table(index='timestamp', columns='username', values='event_type', aggfunc=lambda x: len(x.unique()))

# Create the heatmap using seaborn
plt.figure(figsize=(12, 8))
sns.heatmap(log_matrix, cmap='coolwarm', linewidths=.5)

# Set the plot title and axis labels
plt.title('User Activity Heatmap')
plt.xlabel('Username')
plt.ylabel('Timestamp')

# Show the plot
plt.show()
```
